## Laptop Price Analysis

 The project titled “Laptop Price Analysis Dashboard” aims to analyze how different laptop specifications influence their market price and to visualize these insights through an interactive dashboard. The primary objective is to identify key factors affecting laptop pricing, such as brand, processor type, RAM capacity, storage type, screen size, and operating system. This analysis will help consumers make informed purchase decisions and enable retailers to optimize their pricing strategies.



The selected dataset — Laptop Price Dataset from Kaggle — contains approximately 1,300 records and 12 diverse features, including both numerical (price, RAM, weight) and categorical (brand, type, processor, OS) variables. The data is clean, well-structured, and publicly available from a credible source, making it suitable for exploratory data analysis and visualization. The domain of this project falls under Sales and E-commerce, a highly relevant field in today’s data-driven business environment. The dataset size, quality, and variety meet the project requirements, providing a strong foundation for deriving meaningful insights and building an analytical dashboard.

## Brand Value and Price Comparison

```python
# Average price by brand

brand_price = df.groupby('Company')['Price'].mean().sort_values()
print(brand_price)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Visualize
plt.figure(figsize=(10,5))
sns.barplot(x=brand_price.index,hue=brand_price.index ,y=brand_price.values, palette='coolwarm')
plt.title("Average Laptop Price by Brand")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

## Does higher RAM or Storage always increase price?

```python
# Average price by RAM size
ram_price = df.groupby('Ram')['Price'].mean().sort_values(ascending=False)
print(ram_price)


```

```python
# Visualize RAM impact
plt.figure(figsize=(8,5))
sns.barplot(x=ram_price.index, y=ram_price.values)
plt.title("Average Price vs RAM")
plt.ylabel("Average Price")
plt.show()
```

#### "From this figure it is clear that Laptop containing Highest Ram  contains highest price"
Higher RAM capacity almost always results in a higher laptop price.
However, the jump in cost from 16GB to 32GB and 64GB is especially steep, indicating that ultra-high RAM configurations target a specialized, high-performance market.

```python
# Average price by Storage
# Simplify storage categories
storage_price = df.groupby('Memory')['Price'].mean().sort_values(ascending=False)
# storage_price = storage_price.head(10)
print(storage_price)




```

Insight

### Brand & Processor Analysis
Simplify the Data (Focus on Top Categories)

Here  only display:

The top 10 brands by average price.

The most common processors (top 5).

```python
# Simplify the dataset
top_brands = df['Company'].value_counts().head(10).index
top_processors = df['Cpu'].value_counts().head(5).index

filtered_df = df[df['Company'].isin(top_brands) & df['Cpu'].isin(top_processors)]

# Recalculate average price
brand_processor_price = filtered_df.pivot_table(values='Price', index='Company', columns='Cpu', aggfunc='mean')

brand_processor_price 
```

```python
# Simplified Heatmap
plt.figure(figsize=(15,8))
sns.heatmap(brand_processor_price, annot=True, fmt=".0f", cmap="YlOrBr", linewidths=0.5, cbar_kws={'label': 'Average Price (₹)'})

plt.title("Average Laptop Price (Top Brands & Processors)", fontsize=14, fontweight='bold')
plt.xlabel("Processor Type", fontsize=12)
plt.ylabel("Brand", fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

Processor Impact:

Laptops with Intel Core i9 and Apple M2/M3 chips are the most expensive on average.

Intel Core i5 and Ryzen 5 models are moderately priced, offering good performance for cost.

Pentium and Celeron processors are associated with the lowest prices, targeting budget users.

Brand-Wise Trends:

Apple consistently dominates the higher price range due to its premium build and M-series chips.

Dell, HP, and ASUS show a wide range of prices depending on processor types.

Acer and Lenovo tend to offer relatively affordable options even for higher processors like i7.

Best Value for Money:

Lenovo with Ryzen 5 or HP with Intel i5 provide a balanced mix of price and performance.

These combinations are ideal for students and professionals looking for performance under budget.

High-End Segment:

Apple (M2/M3) and Dell (i9/i7) target the premium user group—creators and developers requiring top performance.

Budget Segment:

Acer (i3) and HP (Pentium/Celeron) dominate the lower price band, appealing to basic users or students.

### Feature Correlation with Price

```python
# Clean and convert RAM (e.g., '8GB' -> 8)
df['Ram'] = df['Ram'].str.replace('GB', '').astype(float)

# Extract storage (Memory) numeric value (e.g., '512GB SSD' -> 512)
df['Storage_GB'] = df['Memory'].str.extract(r'(\d+)').astype(float)

# Extract CPU speed (e.g., 'Intel Core i5 7200U 2.5GHz' -> 2.5)
df['Cpu_speed_GHz'] = df['Cpu'].str.extract(r'(\d+\.\d+)').astype(float)
```

```python
numeric_features = ['Ram', 'Storage_GB', 'Screen_size(Inches)', 'Cpu_speed_GHz', 'Price']
corr = df[numeric_features].corr()
print(corr['Price'].sort_values(ascending=False))
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap with Price", fontsize=14, fontweight='bold')
plt.show()
```

Insights from this heat map

RAM (0.74) shows the strongest positive correlation with laptop price — meaning laptops with higher RAM tend to be significantly more expensive.

CPU Speed (0.43) also has a moderate positive correlation, suggesting faster processors contribute to higher prices.

Storage Capacity (0.30) has a weak but positive correlation, indicating that while larger storage adds value, it’s not as influential as RAM or CPU speed.

Screen Size (0.07) shows an almost negligible relationship with price, implying display size doesn’t heavily affect pricing.

Overall, performance-related features (RAM and CPU speed) have a much greater impact on price than physical characteristics like screen size or storage.

## 2.Impact of RAM and Storage on Price

### RAM has a strong positive impact on price.

As RAM increases from 2GB → 4GB → 8GB → 16GB → 32GB → 64GB, the average price rises sharply.

Laptops with 16GB or more RAM are considerably more expensive and typically target gamers, designers, or developers.

This confirms that RAM is one of the most influential features in determining laptop cost.

### Storage also influences price—but with some variations.

Laptops with SSD-based storage (especially 512GB or 1TB) are significantly pricier than HDD-only models.

However, larger HDDs (e.g., 2TB) may not always increase the price substantially, indicating that type of storage (SSD vs HDD) matters more than just capacity.

### Combined Effect:

Models with both high RAM and large SSDs command premium prices.

The performance-oriented configurations (e.g., 16GB RAM + 512GB SSD) offer top speed but are less cost-efficient compared to mid-tier configurations (8GB RAM + 256GB SSD).

### Exceptions:

Some budget laptops offer higher storage (e.g., 1TB HDD) but still remain low-priced because HDDs are slower and cheaper.

Thus, more storage doesn’t always mean higher price — the type and performance of storage are crucial factors.


#### Laptops with more RAM usually cost more — there’s a clear link between RAM size and price.
#### However, having more storage doesn’t always mean a higher price. The type of storage matters more — SSDs are usually pricier than HDDs, even with less space.
